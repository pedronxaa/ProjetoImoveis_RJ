```python
db['Bairro'] = db['Local'].str.extract(r'em\s+([^,]+),')

db['Aluguel(R$)'] = db['Preço'].str.extract(r'R\$\s*([\d\.]+(?:\/(?:mês|dia))?)')
db['IPTU(R$)'] = db['Preço'].str.extract(r'IPTU\s*R\$\s*([\d\.]+)')
db['Condomínio(R$)'] = db['Preço'].str.extract(r'Cond\.\s*R\$\s*([\d\.]+)')

db['Área(m²)'] = db['Área(m²)'].str.split()
db['Área(m²)'] = db['Área(m²)'].str[-2]

db['Quartos'] = db['Quartos'].str.split()
db['Quartos'] = db['Quartos'].str[-1]

db['Banheiros'] = db['Banheiros'].str.split()
db['Banheiros'] = db['Banheiros'].str[-1]

db['Vagas'] = db['Vagas'].str.split()
db['Vagas'] = db['Vagas'].str[-1]

db['Zona'] = db['Bairro'].apply(lambda x: bairro_zona_map.get(x, 'Outra'))
db.drop(columns=['Local', 'Preço'], inplace=True)
```

### Tratamento de Dados

```python
#Tratamento (remoção) dos aluguéis diários
mensais = db_filtrado[db_filtrado['Aluguel(R$)'].str[-1]!='a']
db_filtrado = mensais
valor = db_filtrado['Aluguel(R$)'].str.replace('/mês', '', regex=False)
db_filtrado['Aluguel(R$)'] = valor
db_filtrado
```

```python
valor_puro = valor.str.replace('.', '', regex=False)
db_filtrado['Aluguel(R$)'] = valor_puro
db_filtrado['Aluguel(R$)'] = db_filtrado['Aluguel(R$)'].astype(float)

cond = db_filtrado ['Condomínio(R$)'].str.replace('.', '', regex=False)
db_filtrado ['Condomínio(R$)'] = cond
db_filtrado['Condomínio(R$)'] = db_filtrado['Condomínio(R$)'].astype(float)

iptu = db_filtrado['IPTU(R$)'].str.replace('.', '', regex=False)
db_filtrado['IPTU(R$)'] = iptu
db_filtrado['IPTU(R$)'] = db_filtrado['IPTU(R$)'].astype(float)

```

```python
areas = []
for valor_str in db_filtrado['Área(m²)']:
    if '-' in valor_str:
        partes = [float(val.strip()) for val in valor_str.split('-')]
        areas.append(np.mean(partes))
    else:
        areas.append(float(valor_str))
db_filtrado['Área(m²)'] = areas

quartos = []
for valor_str in db_filtrado['Quartos']:
    if pd.isna(valor_str):
        quartos.append(np.nan) 
        continue

    if '-' in valor_str:
        partes = [float(val.strip()) for val in valor_str.split('-')] 
        quartos.append(np.mean(partes))
    else: 
        quartos.append(float(valor_str.strip())) 

banheiros = []
for valor_str in db_filtrado['Banheiros']:
    if pd.isna(valor_str):
        banheiros.append(np.nan) 
        continue

    if '-' in valor_str:
        partes = [float(val.strip()) for val in valor_str.split('-')] # Usar float é mais seguro para média
        banheiros.append(np.mean(partes))
    else: 
        banheiros.append(float(valor_str.strip())) # Usar float é mais flexível

vagas = []
for valor_str in db_filtrado['Vagas']:
    if pd.isna(valor_str):
        vagas.append(np.nan) 
        continue

    if '-' in valor_str:
        partes = [float(val.strip()) for val in valor_str.split('-')] # Usar float é mais seguro para média
        vagas.append(np.mean(partes))
    else: 
        vagas.append(float(valor_str.strip())) #
db_filtrado['Área(m²)'] = areas
db_filtrado['Quartos'] = quartos
db_filtrado['Banheiros'] = banheiros
db_filtrado['Vagas'] = vagas
```

#### Tratando Dados faltantes

```python
db_filtrado.shape
db_filtrado.isna().sum()
```

```python
#Os apartamentos com aluguéis faltantes, também possuem IPTU e Condomínio faltantes.(MAR)
db_filtrado[db_filtrado['Aluguel(R$)'].isna()] 
db_filtrado['Aluguel(R$)'] = db_filtrado['Aluguel(R$)'].fillna(np.nanmedian(db_filtrado['Aluguel(R$)']))
```

```python
#Condomínio e IPTU possuem muitos dados faltante por não terem sido informados no anuncio, a maioria deles provavelmente por erro do usuário.
db_filtrado[db_filtrado['Condomínio(R$)'].isna()]
db_filtrado[db_filtrado['IPTU(R$)'].isna()]

db_filtrado['Condomínio(R$)'] = db_filtrado['Condomínio(R$)'].fillna(np.nanmedian(db_filtrado['Condomínio(R$)']))
db_filtrado['IPTU(R$)'] = db_filtrado['IPTU(R$)'].fillna(np.nanmedian(db_filtrado['IPTU(R$)']))
```

```python
#Dados faltantes na coluna "Quartos" sempre acontece quando o apartamento é muito pequeno, provavelmente por ser uma kitnet, substituiremos por 1 então.
db_filtrado[db_filtrado['Quartos'].isna()]

db_filtrado['Quartos'] = db_filtrado['Quartos'].fillna(1)
```

```python
#são poucos os imóveis com dados faltantes de banheiro e ocorre por erro de digitação, então podemos remover esses
db_filtrado[db_filtrado['Banheiros'].isna()]

db_filtrado.dropna(subset = ['Banheiros'], inplace = True)
```

```python
#QUando o número de vagas de garagem não é informado, geralmente é quando aquele apartamento não oferece nenhuma vaga
db_filtrado[db_filtrado['Vagas'].isna()]

db_filtrado['Vagas'] = db_filtrado['Vagas'].fillna(0)
```

```python
#imóveis sem link são frutos de erro na extração de dados, iremos removê-los
db_filtrado[db_filtrado['Link'].isna()]

db_filtrado.dropna(subset = ['Link'], inplace = True)
```
